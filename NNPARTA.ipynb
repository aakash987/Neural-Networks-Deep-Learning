```python
#A. Split the data into X & Y. [1 Marks]
#B. Split the data into train & test with 70:30 proportion.[1 Marks]


from sklearn.model_selection import train_test_split

# A. Splitting the data into X and Y
X = df.drop(columns=['Signal_Strength'])  # Features
y = df['Signal_Strength']  # Target variable

# B. Splitting the data into train and test sets with a 70:30 proportion
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Display the shapes of the train and test sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# C. Print shape of all the 4 variables and verify if train and test data is in sync. [1 Marks]

print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)

# Verify if train and test data are in sync
train_sync = len(X_train) == len(y_train)
test_sync = len(X_test) == len(y_test)

if train_sync and test_sync:
    print("Train and test data are in sync.")
else:
    print("Train and test data are not in sync.")
```

```python
#D. Normalise the train and test data with appropriate method. [2 Marks]

from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the training data
X_train_normalized = scaler.fit_transform(X_train)

# Transform the test data
X_test_normalized = scaler.transform(X_test)

# Print the shapes of the normalized data
print("Shape of X_train_normalized:", X_train_normalized.shape)
print("Shape of X_test_normalized:", X_test_normalized.shape)
```

```python
#E. Transform Labels into format acceptable by Neural Network [2 Marks]

from sklearn.preprocessing import OneHotEncoder

# Initialize the OneHotEncoder
encoder = OneHotEncoder()

# Fit and transform the training labels
y_train_encoded = encoder.fit_transform(y_train.values.reshape(-1, 1)).toarray()

# Transform the testing labels
y_test_encoded = encoder.transform(y_test.values.reshape(-1, 1)).toarray()

# Print the shapes of the encoded labels
print("Shape of y_train_encoded:", y_train_encoded.shape)
print("Shape of y_test_encoded:", y_test_encoded.shape)
```

```python
#A. Design a Neural Network to train a classifier. [3 Marks]

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the input shape
input_shape = X_train_normalized.shape[1]

# Initialize the model
model = Sequential()

# Add the input layer
model.add(Dense(64, activation='relu', input_shape=(input_shape,)))

# Add one hidden layer
model.add(Dense(32, activation='relu'))

# Add the output layer
# Since we have 6 classes, we use softmax activation for multi-class classification
model.add(Dense(6, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',  # for multi-class classification
              metrics=['accuracy'])

# Print the model summary
model.summary()
```

```python
# B. Train the classifier using previously designed Architecture [2 Marks]

history = model.fit(X_train_normalized, y_train_encoded, epochs=50, batch_size=32, validation_split=0.1, verbose=1)
```

```python
#Plot 2 separate visuals. [3 Marks]
import matplotlib.pyplot as plt

# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()


```

```python
# C. Plot 2 separate visuals. [3 Marks]


plt.figure(figsize=(10, 5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
#D. Design new architecture/update existing architecture in attempt to improve the performance of the model. [2 Marks]


from tensorflow.keras.layers import Dropout

# Initialize the model
model_new = Sequential()

# Add the input layer
model_new.add(Dense(128, activation='relu', input_shape=(input_shape,)))
model_new.add(Dropout(0.2))  # Adding dropout regularization

# Add a hidden layer
model_new.add(Dense(64, activation='relu'))
model_new.add(Dropout(0.2))  # Adding dropout regularization

# Add another hidden layer
model_new.add(Dense(32, activation='relu'))
model_new.add(Dropout(0.2))  # Adding dropout regularization

# Add the output layer
model_new.add(Dense(6, activation='softmax'))

# Compile the model
model_new.compile(optimizer='adam',
              loss='categorical_crossentropy',  # for multi-class classification
              metrics=['accuracy'])

# Print the model summary
model_new.summary()
```

```python
# D. Design new architecture/update existing architecture in attempt to improve the performance of the model. [2 Marks]



from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Define the model architecture
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_normalized.shape[1],)),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(6, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print the model summary
model.summary()

```

```python
#E. Plot visuals as in Q3.C and share insights about difference observed in both the models. [3 Marks]

import matplotlib.pyplot as plt

# Train the original model and capture history
history_original = model.fit(X_train_normalized, y_train_encoded, epochs=50, batch_size=32, validation_split=0.1, verbose=1)

# Train the improved model and capture history
history_improved = model_new.fit(X_train_normalized, y_train_encoded, epochs=50, batch_size=32, validation_split=0.1, verbose=1)

# Plot training & validation accuracy values
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_original.history['accuracy'], label='Original Model - Training Accuracy', color='blue')
plt.plot(history_original.history['val_accuracy'], label='Original Model - Validation Accuracy', color='orange')
plt.plot(history_improved.history['accuracy'], label='Improved Model - Training Accuracy', color='green')
plt.plot(history_improved.history['val_accuracy'], label='Improved Model - Validation Accuracy', color='red')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history_original.history['loss'], label='Original Model - Training Loss', color='blue')
plt.plot(history_original.history['val_loss'], label='Original Model - Validation Loss', color='orange')
plt.plot(history_improved.history['loss'], label='Improved Model - Training Loss', color='green')
plt.plot(history_improved.history['val_loss'], label='Improved Model - Validation Loss', color='red')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
